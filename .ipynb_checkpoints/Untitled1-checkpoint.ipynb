```python
import pandas as pd 
import numpy as np
import re
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data=pd.read_csv('tmdb_5000_movies.csv')
```

```python
def moviewords(col):
    new_col=[]
    for i in data[col]:
        i=str(i)
        i=i.lower()
        i=re.sub('[0-9]', '', i)
        i=re.sub('name|id', '', i)
        i=re.sub('{"": , "": "|"}', '', i)
        i=re.sub('\\[|\\]', '', i)
        i=re.sub('{"": "|", "": }', '', i)
        new_col.append(i)
    return new_col
```

```python
modify=data[['id', 'title', 'overview', 'popularity', 'budget', 'revenue',
             'runtime', 'vote_average', 'vote_count']]
```

```python
modify
```

```python
modify['genres']=moviewords('genres')
modify['keywords']=moviewords('keywords')
modify['production_companies']=moviewords('production_companies')
modify['features']=modify['keywords']+modify['production_companies']+modify['genres']+modify['title']
modify['index']=list(range(0, len(modify)))
modify=modify.drop_duplicates()
```

```python
print(modify.dtypes)
```

```python
modify
```

```python
modify=modify.fillna('')
```

```python
def movie_similar(movie_names):
    movie_1=modify[modify['title']==movie_names[0]]
    movie_2=modify[modify['title']==movie_names[1]]
    movies_df=pd.concat([movie_1, movie_2], axis=0)
    count_matrix=CountVectorizer().fit_transform(movies_df['features'])
    cosine_sim=np.round(cosine_similarity(count_matrix), 3)
    sim=cosine_sim[0,1]
    print(sim)
    return sim
```

```python
def movie_recom(movie_names):
    count_matrix=CountVectorizer().fit_transform(modify['features'])
    cosine_sim=np.round(cosine_similarity(count_matrix), 3)
    def movie_index(title):
        return modify.index[modify['title']==title].values[0]
    for movie_name in movie_names:
        index=movie_index(movie_name)
        similar_movies=list(enumerate(cosine_sim[index]))
        similar_movies=pd.DataFrame(similar_movies, columns=['index', 'similarity'])
        similar_movies=pd.merge(similar_movies, modify[['index', 'title']], on='index')
        similar_movies=similar_movies.sort_values(by='similarity', ascending=False)
        print(similar_movies[1:10])
        print('\n')
        similar_movies = pd.merge(similar_movies[1:10], modify[['title','popularity','vote_average']], on="title")
    return similar_movies
```

```python
movie=['Spider-Man', 'The Avengers']
movie_similar(movie)
movie_recom(movie)
```
